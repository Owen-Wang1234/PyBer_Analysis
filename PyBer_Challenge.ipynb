### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import os

# Files to Load
city_data_to_load = os.path.join("Resources", "city_data.csv")
ride_data_to_load = os.path.join("Resources", "ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# 1. Get the total rides for each city type
rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
rides_by_type
```

```python
# 2. Get the total drivers for each city type
drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
drivers_by_type
```

```python
# 3. Get the total amount of fares for each city type
total_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_type
```

```python
# 4. Get the average fare per ride for each city type. 
fare_per_ride = total_fares_by_type / rides_by_type
fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
fare_per_driver = total_fares_by_type / drivers_by_type
fare_per_driver
```

```python
# 6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.concat([rides_by_type, drivers_by_type, total_fares_by_type,
                              fare_per_ride, fare_per_driver], axis = 1)

# Rename the columns
pyber_summary_df = pyber_summary_df.rename(columns = {"ride_id" : "Total Rides", "driver_count" : "Total Drivers",
                                                      "fare" : "Total Fares", 0 : "Average Fare per Ride",
                                                      1 : "Average Fare per Driver"})
pyber_summary_df
```

```python
# 7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
# 8. Format the columns.
# Drawing from https://www.stackoverflow.com/questions/209375328/
# how-to-display-pandas-dataframe-of-floats-using-a-format-string-for-columns
pyber_summary_df = pyber_summary_df.style.format({"Total Fares" : "${:,.2f}",
                                                  "Average Fare per Ride" : "${:,.2f}",
                                                  "Average Fare per Driver" : "${:,.2f}"})

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_type_time = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
fares_type_time
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_type_time = fares_type_time.reset_index()
fares_type_time
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_pivot = fares_type_time.pivot(index = "date", columns = "type", values = "fare")
fares_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fares_2019_df = fares_pivot.loc["2019-01-01" : "2019-04-28"]
fares_2019_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fares_2019_df.index = pd.to_datetime(fares_2019_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fares_2019_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_2019_df = fares_2019_df.resample("W").sum()
weekly_fares_2019_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Set up the line chart.
fares_2019 = weekly_fares_2019_df.plot(figsize = (24, 8))

# Add the title.
plt.title("Total Fare by City Type")

# Set the axis labels.
plt.xlabel("")
plt.ylabel("Fare ($USD)")

# Save an image of the chart and print it to the screen.
chart_to_save = os.path.join("analysis", "PyBer_fare_summary.png")
plt.savefig(chart_to_save)
plt.show()
```
