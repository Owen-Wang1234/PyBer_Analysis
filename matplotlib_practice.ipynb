```python
# Keep the plots displayed inline with "magic."
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the line plot.
plt.plot(x_axis, y_axis)
```

```python
# Create the same plot the Object-Oriented way.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create an annotated plot with a label for the legend.
plt.plot(x_axis, y_axis, marker = '*', color = 'blue', linewidth = 2, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title('PyBer Fare by Month')

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the same annotated plot the Object-Oriented way.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = 'D', color = 'green', linewidth = 2, label = 'Boston')

# Create labels for the x and y axes.
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)

# Create a title.
ax.set_title('PyBer Fare by Month')

# Add a grid.
ax.grid()

# Add the legend.
ax.legend()
```

```python
# Create the bar plot.
plt.bar(x_axis, y_axis)
```

```python
# Create an annotated bar plot.
plt.bar(x_axis, y_axis, color = 'green', label = 'Boston')

# Create labels for the x and y axis.
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Create a title.
plt.title('PyBer Fare by Month')

# Add the legend.
plt.legend()
```

```python
# Create a horizontal bar plot with the date in the right order.
plt.barh(x_axis, y_axis, color = 'magenta', label = 'Boston')
plt.gca().invert_yaxis()

# Create labels for the x and y axis.
plt.xlabel('Fare($)')
plt.ylabel('Date')

# Create a title.
plt.title('PyBer Fare by Month')

# Add the legend.
plt.legend()
```

```python
# Create the same bar plot the Object-Oriented way.
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the horizontal bar plot the Object-Oriented way.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = 'cyan', label = 'Chicago')
ax.invert_yaxis()

# Create labels for the x and y axes.
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')

# Create a title.
ax.set_title('PyBer Fare by Month')

# Add the legend.
ax.legend()
```

```python
# Create a scatter plot.
plt.scatter(y_axis, x_axis, color = 'red', label = 'Chicago')
plt.gca().invert_yaxis()

# Create labels for the x and y axis.
plt.xlabel('Fare($)')
plt.ylabel('Date')

# Set the x limit between 0 and 50.
plt.xlim(0, 50)

# Create a title.
plt.title('PyBer Fare by Month')

# Add the legend.
plt.legend()
```

```python
# Create a bubble chart.
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# Set up larger markers; magnified by 3x.
y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data * 3)

# Try the bubble chart now.
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# Same magnification with list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter plot the Object-Oriented way
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble chart the Object-Oriented way
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color = 'skyblue', alpha = 0.8,
           linewidths = 2, edgecolors = 'black', label = 'Boston')
ax.invert_yaxis()

# Create labels for the x and y axes.
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')

# Set the x limit between 0 and 50.
ax.set_xlim(0, 50)

# Create a title.
ax.set_title('PyBer Fare by Month')

# Add the legend.
ax.legend()
```

```python
# Create a pie chart.
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
# Create an exploded pie chart.

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow",
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Set up the explode parameters
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize = (8, 8))
plt.pie(y_axis, explode = explode_values, colors = colors,
        labels = x_axis, autopct = '%.1f%%')
plt.show()
```

```python
# Same pie chart the Object-Oriented way

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow",
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Set up the explode parameters
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize = (8, 8))
ax.pie(y_axis, explode = explode_values, colors = colors,
       labels = x_axis, autopct = '%.1f%%', shadow = True,
       startangle = 90, counterclock = False)
plt.show()
```
